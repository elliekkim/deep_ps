````python
# Open up this csv file and visualize it:

import matplotlib.pyplot as plt
import pandas as pd

# Read in the data
data = pd.read_csv('2023_nc_july_7pm_avg.csv')

data.head()
````

````python
# Visualize temp_eve against lat_m, lon_m as a 2d plot:

plt.scatter(data['lon_m'], data['lat_m'], c=data['temp_eve'])
plt.colorbar()
plt.xlabel('Latitude')
plt.ylabel('Longitude')
plt.title('Temperature at 7pm')
plt.show()
````
